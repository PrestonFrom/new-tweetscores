```r
options(repr.plot.width=12, repr.plot.height=6.75)
```

```r
vf <- read_tsv("../data/thetas.tsv", col_types = 'cnfnnnnn')
```

```r
vf <- filter(vf, follows  > 5)
```

```r
ggplot(vf, aes(x=scaled_theta, fill=party)) + geom_density() + facet_wrap(vars(party)) + scale_fill_manual(values=c("firebrick", "goldenrod", "dodgerblue")) + theme_bw()
```

```r
ggplot(vf, aes(x=scaled_theta_2, fill=party)) + geom_density() + facet_wrap(vars(party)) + scale_fill_manual(values=c("firebrick", "goldenrod", "dodgerblue")) + theme_bw()
```

```r
ggplot(vf, aes(y=follows, x=scaled_theta, color=party)) + geom_point(alpha=0.3) + scale_color_manual(values=c("firebrick", "goldenrod", "dodgerblue")) + theme_bw()
```
